```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
```

```python
def getplots(algorithms, dsets):
    i = 1
    plt.figure(figsize=(15,10))
    m,n =len(algorithms),len(dsets)
    for alg in algorithms:
        for dset in dsets:
            plt.subplot(m,n,i)
            alg = alg
            predi = alg.fit_predict(dset)
            plt.scatter(dset['x'],dset['y'], c=predi, cmap='rainbow')
            i+=1
    plt.tight_layout
    plt.show()
```

```python
getplots([KMeans(), AgglomerativeClustering(), DBSCAN(eps=1.2, min_samples=3)], data_sets)
```

## Playing with Cosine Similarity

```python
from functools import reduce
```

Defining the dot product $$v_1 \circ v_2$$

```python
RANDOM_STATE = 4
np.random.seed = RANDOM_STATE
v1 = np.random.randint(0,10,5)
v2 = np.random.randint(0,10,5)
print('v1 = {}, v2 = {}'.format(v1,v2))
print('v1 * v2 = {}'.format(reduce(lambda x,y : x +' + '+y,[str(x) +'*'+ str(y) for x,y in zip(v1,v2)])))
```

The cosine similarity between two vectors is the cosine of the angle between them.

For example, parallel verctors have an angle of 0, and Cos(0) = 1.

PARALLEL vectors have cosine_similarity(v1,v2) = 1

On the other hand, perpendicular vectors are at 90 degrees to each other, and hace cosine similarity = cos(90) = 0

The definition is:

Cosine similarity = $cos(\theta_{v_1,v_2}) = \frac{v_1*v_2}{|v_1||v_2|}$ where $\theta$ is the angle between the vectors.

```python
from sklearn.metrics.pairwise import cosine_similarity,cosine_distances
# Note, cosine_similarity likes 
```

```python
cosine_similarity([v1,v2])
```

The output above is the cosine similarity between v1 and itself (always 1), v1 and v2, v2 and v1 and then v2 and v2 (always 1)

## Tokenizing Text and using Tf-idf
http://scikit-learn.org/stable/modules/feature_extraction.html

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
# CountVectorizer does what it says on the tin
text1 = """Twas brillig, and the slithy toves
Did gyre and gimble in the wabe.
All mimsy were the borogoves
And the mome raths outgrabe."""

text2 = """Beware the Jabberwock, my son!
The jaws that bite, the claws that catch!
Beware the Jubjub bird, and shun
The frumious Bandersnatch!"""


text3 = """He took his vorpal sword in his hand,
Long time the manxome foe he sought --
So rested he by the Tumtum tree,
And stood awhile in thought."""

text4 = """And, as in uffish thought he stood,
The Jabberwock, with eyes of flame,
Came whiffling through the tulgey wood,
And burbled as it came!"""

text5 = """One, two! One, two! And through and through!
The vorpal blade went snicker-snack!
He left it dead, and with its head
He went galumphing back."""

text6 = """And, hast thou slain the Jabberwock?
Come to my arms, my beamish boy!
O frabjous day! Callooh! Callay!
He chortled in his joy."""

text7 = """`Twas brillig, and the slithy toves
Did gyre and gimble in the wabe.
All mimsy were the borogoves,
And the mome raths outgrabe."""

corpus_df = pd.Series([text1,text2,text3,text4,text5,text6,text7])
corpus_df.head()
```

```python
count_vectorizer = CountVectorizer() # you can set many options here
counts = count_vectorizer.fit_transform(corpus_df)
counts 
```

Note that the output is a Sparce Matrix. This is a much better way to represent this kind of matrices for memory consumption reasons. If this were a big courpus we could easily have 1m+ features, and most entries would be 0, so keeing it "sparce" saves a lot of memory.

Since this is a small corpus, we can look at it densly:

```python
dense_counts = counts.todense()
dense_counts
```

```python
# This is easyer to look at as a DataFrame
df_using_count_vectorizer = pd.DataFrame(dense_counts)
df_using_count_vectorizer
```

This is a tally of words as seen in the 7 texts of the corpus.
The next step is not important for running models, but very interesing for EDA: looking at what are the words.

This is stored as the "feature_names" in the count vectorizer we just used

```python
count_vectorizer.get_feature_names()
```

```python
df_using_count_vectorizer.columns = count_vectorizer.get_feature_names()
df_using_count_vectorizer
```

Cool.

Tf-idf does 'the same thing' (those are air quotes to indicate that it's not quite the same thing).

Tf-idF tallies the words but then alters the "score" of each word. If the word is very frequent in the current text, the score goes up; if it is very frecuent in all texts, the word goes down.

Then all words in the text are 're-scored' so that the euclidean length of the text vector is 1.

In short: It tallies words in a clever way.

```python
tf_idf = TfidfVectorizer()
df_using_tf_idf = pd.DataFrame(tf_idf.fit_transform(corpus_df).todense(),
                               columns=tf_idf.get_feature_names())
df_using_tf_idf
```

It may be good to compare CountVectorizer and Tf-idf. Lets look at the words 'jaws' and 'Jabberwock'

```python
# we see that the word 'jaws' appears only in text2 and only once.
df_using_count_vectorizer.jaws
```

```python
# The Tf-idf score for jaws in text two is slightly less than 
# one because of the normalization 
# to vector length 1
df_using_tf_idf.jaws
```

```python
# We can see that jabberwock is much more common, being present in 4/7 texts in our corpus
df_using_count_vectorizer.jabberwock
```

```python
# And because of that the values for jabberwock are much lower 
# Note that jaw and jabberwock both featured once in text2, but jabberwock hot a lower score
df_using_tf_idf.jabberwock
```

## Using cosine similarity for clustering

```python
# There are ways of setting up a precomputed cosine similarty
# matrix as a "distance" meassure between texts. However, life is easy,
# sklearn's clustering algorithms have 'cosine similarity' as a setting 

# Say I want 2 clusters:
cos_sim_agglomerative_clustering = AgglomerativeClustering(n_clusters=2,
                                                           affinity='cosine',
                                                           linkage='average')
```

```python
cos_sim_agglomerative_clustering.fit_predict(df_using_tf_idf)
```

Here we see the clustering algorithm give us our requested 2 clusters. 

The theory is that texts 1,2, and 3 have a lot in common with each other, and texts 3-6 have a lot in common amongst themseves. This may or may not be true, but that is what we hoped.

```python
cos_sim_DBSCAN_clustering = DBSCAN(metric='cosine',eps = .8, min_samples=2)
cos_sim_DBSCAN_clustering.fit_predict(df_using_tf_idf)
```

DBSCAN with the settings we passed tells us texts 1,2 and 7 have a lot in common with each other, and that texts 3-6 are all outliers (special snowflakes with little in common to each other other texts)
